## Step 4 : Exploratory Data Analysis (EDA)

- Summary Statistics

```python
def summary_statistics(df: pd.DataFrame) -> None:
    """Display summary statistics of numerical and categorical features."""
    print("Summary Statistics for Numerical Features:")
    print(df.describe())

    print("\nSummary Statistics for Categorical Features:")
    print(df.describe(include=['O']))
```

```python
summary_statistics(df)
```

- Distribution of Numerical Features

```python
# Step 2: Distribution of Numerical Features
def plot_numerical_distributions(df: pd.DataFrame, columns: list) -> None:
    """Plot distributions of the numerical features."""
    for column in columns:
        fig = px.box(df, x=column)
        fig.show()
```

```python
numerical_columns = ['Quantity', 'UnitPrice', 'CustomerID']  # Adjust as needed
plot_numerical_distributions(df, numerical_columns)
```

- Distribution of Categorical Features

```python
def plot_categorical_distributions(df: pd.DataFrame, column: str) -> None:
    """Plot the distribution of a categorical feature."""
    fig = px.histogram(df, x=column, title=f'Distribution of {column}',
                       color_discrete_sequence=px.colors.qualitative.Set2)
    fig.show()
```

```python
# Visualize distributions of 'StockCode' and 'Country'
plot_categorical_distributions(df, 'Country')
plot_categorical_distributions(df, 'StockCode')
```

- Missing Values Analysis

```python
missing_data = df.isnull().sum()
missing_data = missing_data[missing_data > 0]
missing_data
```

```python
def missing_values_analysis(df: pd.DataFrame) -> None:
    """Analyze and visualize missing values in the DataFrame."""
    missing_data = df.isnull().sum()
    missing_data = missing_data[missing_data > 0]

    fig = px.bar(
        x=missing_data.index,
        y=missing_data.values,
        title='Missing Values Analysis',
        labels={'x': 'Features', 'y': 'Number of Missing Values'})
    fig.show()
```

```python
missing_values_analysis(df)
```

- Groupby Analysis

```python
def group_by_analysis(df: pd.DataFrame, groupby_column: str) -> None:
    """Analyze and plot aggregated statistics by category."""
    group_df = df.groupby(groupby_column)['Quantity'].agg(['sum', 'mean', 'count']).sort_values(by='sum', ascending=False).reset_index()

    fig = px.bar(group_df, x=groupby_column, y='sum',
                  title=f'Sum of Quantities by {groupby_column}',
                  labels={groupby_column: groupby_column, 'sum': 'Total Quantity'})
    fig.show()
```

```python
group_by_analysis(df, 'Country')  # Example grouping by Country
```
